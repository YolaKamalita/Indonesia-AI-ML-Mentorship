## Dealing with Missing Data

```python
# Nulls percentage for each row

def percent_missing(df):
    percent_null = df.isnull().sum() / len(df)
    percent_null = percent_null[percent_null>0].sort_values()
    return percent_null
```

# Statistical Modelling

## Create Model

```python
X = final_df.drop(['SalePrice'], axis = 1)
y = final_df['SalePrice']
```

```python
X.head()
```

## Polynomial Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
X_OTHER, X_test, y_OTHER, y_test = train_test_split(X, y, test_size=0.10, random_state=103)
```

```python
X_OTHER.shape
```

```python
y_OTHER.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
train_rmse_errors = []
eval_rmse_errors = []
```

```python
# Polynomial Degree 1

polynomial_converter = PolynomialFeatures(degree=1,include_bias=False)
poly_features = polynomial_converter.fit_transform(X_OTHER)
    
X_train, X_eval, y_train, y_eval = train_test_split(poly_features, y_OTHER, test_size=0.10, random_state=103)
    
model = LinearRegression()
model.fit(X_train,y_train)
    
train_pred = model.predict(X_train)
eval_pred = model.predict(X_eval)
    
train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
eval_RMSE = np.sqrt(mean_squared_error(y_eval,eval_pred))
   
train_rmse_errors.append(train_RMSE)
eval_rmse_errors.append(eval_RMSE)
```

```python
# Polynomial Degree 2

polynomial_converter = PolynomialFeatures(degree=2,include_bias=False)
poly_features = polynomial_converter.fit_transform(X_OTHER)
    
X_train, X_eval, y_train, y_eval = train_test_split(poly_features, y_OTHER, test_size=0.10, random_state=103)
    
model = LinearRegression()
model.fit(X_train,y_train)
    
train_pred = model.predict(X_train)
eval_pred = model.predict(X_eval)
    
train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
eval_RMSE = np.sqrt(mean_squared_error(y_eval,eval_pred))
   
train_rmse_errors.append(train_RMSE)
eval_rmse_errors.append(eval_RMSE)
```

```python
pd.DataFrame([train_rmse_errors, eval_rmse_errors], columns=['Train_RMSE', 'Evaluation_RMSE'])
```

Both polynomial degree 1 and 2 produce bad accuracy.

Hypothesis :
    - Need Feature Selection

```python
# Checking accuracy to Test Set (Polynomial degree 1)

final_poly_converter_1 = PolynomialFeatures(degree=1,include_bias=False)
final_model_1 = LinearRegression()
final_model_1.fit(final_poly_converter_1.fit_transform(X_OTHER),y_OTHER)
```

```python
X_test_poly = final_poly_converter_1.fit_transform(X_test)

test_pred = final_model_1.predict(X_test_poly)
test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
```

```python
test_RMSE
```

```python
# Checking accuracy to Test Set (Polynomial degree 2)

final_poly_converter_2 = PolynomialFeatures(degree=2,include_bias=False)
final_model_2 = LinearRegression()
final_model_2.fit(final_poly_converter_2.fit_transform(X_OTHER),y_OTHER)
```

```python
X_test_poly = final_poly_converter_2.fit_transform(X_test)

test_pred = final_model_2.predict(X_test_poly)
test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
```

```python
test_RMSE
```

## Elastic Net

```python
# Train Validation Test Split

X_OTHER, X_test, y_OTHER, y_test = train_test_split(X, y, test_size=0.1, random_state=103)
X_train, X_eval, y_train, y_eval = train_test_split(X_OTHER, y_OTHER, test_size=0.1, random_state=103)
```

```python
# Data Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_eval = scaler.transform(X_eval)
X_test = scaler.transform(X_test)
```

```python
# Create Model

from sklearn.linear_model import ElasticNet

base_elastic_model = ElasticNet()
```

```python
# Grid Search

from sklearn.model_selection import GridSearchCV

param_grid = {'alpha':[0.1, 1, 5, 10, 25, 50, 75, 100],
              'l1_ratio':[0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99, 1]}

grid_model = GridSearchCV(estimator = base_elastic_model,
                          param_grid = param_grid,
                          scoring = 'neg_root_mean_squared_error',
                          cv = 10,
                          verbose = 2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
grid_model.best_estimator_.coef_
```

```python
# Checking with X_test

y_pred = grid_model.predict(X_test)
test_RMSE = np.sqrt(mean_squared_error(y_test,y_pred))
```
