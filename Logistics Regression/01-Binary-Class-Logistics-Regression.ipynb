# Data

An experiment was conducted on 5000 participants to study the effects of age and physical health on hearing loss, specifically the ability to hear high pitched tones.

Features

- age - Age of participant in years

# Logistics Regression

### Train and Test Split | Scaling ###

```python
X = df.drop('test_result',axis=1)
y = df['test_result']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

### Model ###

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression()
log_model.fit(scaled_X_train,y_train)
```

### Coefficient Interpretation

```python
log_model.coef_
```

```python
X.columns
```

Interpretation:
    - number of people passing the test decreasing when ages increasing
    - number of people passing the test increasing when physical score increasing
    - looking at the magnitude of coefficient, physical score is stronger feature in determining test result

### Model Performance

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
y_pred = log_model.predict(scaled_X_test)
```

```python
# Accuracy Score

accuracy_score(y_test,y_pred)
```

```python
# Confusion Matrix

plot_confusion_matrix(log_model,scaled_X_test,y_test, normalize='all')
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# Classification Matrix Report

print(classification_report(y_test,y_pred))
```

### AUC Curves Evaluation

```python
from sklearn.metrics import precision_recall_curve,plot_precision_recall_curve,plot_roc_curve
```

```python
plot_precision_recall_curve(log_model,scaled_X_test,y_test)
```

```python
plot_roc_curve(log_model,scaled_X_test,y_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# Precision Score

precision_score(y_test,y_pred)
```

```python
# Recall Score

recall_score(y_test,y_pred)
```

```python
# F1 Score

f1_score(y_test,y_pred)
```
