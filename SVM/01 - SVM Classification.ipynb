# SVM Kernel Exploration

```python
y = df['Virus Present']
X = df.drop('Virus Present', axis=1)
```

```python
from sklearn.svm import SVC
```

```python
def plot_svm_boundary(model,X,y):
    
    X = X.values
    y = y.values
    
    # Scatter Plot
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='viridis')
    
    # plot the decision function
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = model.decision_function(xy).reshape(XX.shape)

    # plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    # plot support vectors
    ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100,
               linewidth=1, facecolors='none', edgecolors='k')
    plt.show()
```

## Linear Kernel

```python
linear_model = SVC(kernel='linear', C=1000)
linear_model.fit(X, y)
```

```python
plot_svm_boundary(linear_model,X,y)
```

```python
linear_model = SVC(kernel='linear', C=1)
linear_model.fit(X, y)
```

```python
plot_svm_boundary(linear_model,X,y)
```

```python
linear_model = SVC(kernel='linear', C=0.05)
linear_model.fit(X, y)
```

```python
plot_svm_boundary(linear_model,X,y)
```

## Sigmoid Kernel

## Perform GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.1,1,10,100],
              'kernel':['linear','rbf','sigmoid','poly'],
              'gamma':[0.01,0.05,0.1,1],
              'degree':[2,3]}
grid = GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_model = SVC(C=0.1, degree=2, gamma=0.01, kernel='linear')
```

```python
best_model.fit(X,y)
```

```python
plot_svm_boundary(best_model,X,y)
```

## Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=101)
```

```python
best_model.fit(X_train,y_train)
```

```python
plot_svm_boundary(best_model,X,y)
```

```python
y_pred = best_model.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
